### Seaborn Styling,Figure Style and Scale

### 1. Built-in Themes
#### This snippet sets a built-in Seaborn theme ("darkgrid") and then creates a strip plot of total bill by day using the tips dataset.The built-in theme automatically applies background styling and grid lines to your plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Apply the "darkgrid" theme and create a strip plot
sns.set_style("darkgrid")
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot with Darkgrid Theme")
plt.show()
```

### 2. Background Color
#### These snippets demonstrate changing the background style of your plot using Seaborn's style settings. One example uses the "dark" style, and another uses the "ticks" style for a different background appearance.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Use the "dark" style background
sns.set_style("dark")
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot with Dark Background")
plt.show()

# Use the "ticks" style background
sns.set_style("ticks")
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot with Ticks Background")
plt.show()
```

### 3. Grids
#### These examples show how to apply grid styles to your plots using Seaborn. Using "whitegrid" adds a white background with grid lines, while "white" applies a plain white background.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Use "whitegrid" style to display a grid on a white background
sns.set_style("whitegrid")
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot with Whitegrid")
plt.show()

# Use "white" style for a clean white background without grids
sns.set_style("white")
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot with White Background")
plt.show()
```

### 4. Despine
#### These snippets demonstrate how to remove (or "despine") the top and right spines from the plot. One removes all extra spines, and another removes just the left and bottom spines for a cleaner look.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Remove all spines from a white-themed plot
sns.set_style("white")
sns.stripplot(x="day", y="total_bill", data=tips)
sns.despine()
plt.title("Strip Plot with Despine")
plt.show()

# Remove only the left and bottom spines from a whitegrid-themed plot
sns.set_style("whitegrid")
sns.stripplot(x="day", y="total_bill", data=tips)
sns.despine(left=True, bottom=True)
plt.title("Strip Plot with Partial Despine")
plt.show()
```

## 5. Scaling Figure Styles for Different Mediums
### 5A. Scaling Plots for Paper Context
#### This snippet sets the context to "paper" (smallest) and creates a strip plot.It adjusts elements like font size and line width suitable for printed materials.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Set style and context for a small (paper) format
sns.set_style("ticks")
sns.set_context("paper")
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot in Paper Context")
plt.show()
```

### 5B. Scaling Plots for Poster Context
### This snippet sets the context to "poster" (largest) and creates a strip plot. It increases the size of plot elements to be easily visible in large-format presentations.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Set style and context for a large (poster) format
sns.set_style("ticks")
sns.set_context("poster")
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot in Poster Context")
plt.show()
```

### 6. Scaling Fonts and Line Widths
#### These examples illustrate how to fine-tune the appearance by adjusting font scale and grid line widths. One example reduces the font scale and line width, while the other increases the grid line width for emphasis

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Set context with smaller font and thinner grid lines
sns.set_context("poster", font_scale=0.5, rc={"grid.linewidth": 0.6})
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot with Small Font & Thin Grid Lines")
plt.show()

# Set context with default poster font and thicker grid lines
sns.set_context("poster", font_scale=1, rc={"grid.linewidth": 5})
sns.stripplot(x="day", y="total_bill", data=tips)
plt.title("Strip Plot with Default Font & Thick Grid Lines")
plt.show()
```

### 7. The RC Parameter
#### This snippet shows how to check the current plotting context using the RC parameter. It sets a style and context, plots a strip plot, and then retrieves the current plotting context for inspection.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Set style and context, then create a strip plot
sns.set_style("ticks")
sns.set_context("poster")
sns.stripplot(x="day", y="total_bill", data=tips)

# Retrieve and print the current plotting context parameters
current_context = sns.plotting_context()
print("Current Plotting Context:", current_context)

plt.title("Strip Plot with RC Parameter")
plt.show()
```

### 8. Temporarily Setting the Plot Style
#### This snippet demonstrates how to temporarily set the plot style using a context manager. It creates a subplot with a specific axes style for the first plot, then resets for the next subplot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")

# Temporarily set the axes style to "white" for the first subplot
with sns.axes_style("white"):
    plt.subplot(211)
    sns.swarmplot(x="species", y="petal_length", data=iris)
    plt.title("Swarmplot (Iris) with White Axes Style")

# Use default style for the second subplot
plt.subplot(212)
sns.violinplot(x="total_bill", data=tips)
plt.title("Violin Plot (Tips)")
plt.tight_layout()
plt.show()
```

## 9. Setting Figure Size in Seaborn
### 9A. Custom Figure Size for a Swarmplot
#### This snippet creates a figure with a specified size and plots a swarmplot using the iris dataset. It demonstrates how to control the overall figure dimensions for better layout control.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a figure and axes object with a custom size
fig, ax = plt.subplots(figsize=(10, 4))
sns.swarmplot(x="species", y="petal_length", data=iris, ax=ax)
plt.title("Swarmplot (Iris) with Custom Figure Size")
plt.show()
```

### 9B. Factor Plot (Categorical Bar Plot) with Custom Size
#### This snippet uses a factor plot (now known as catplot) to visualize survival rates by class and sex in the Titanic dataset. It sets the figure size using the size and aspect parameters for flexible layout control.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Create a factor plot (catplot) with custom size and aspect ratio
g = sns.catplot(x="class", y="survived", hue="sex", data=titanic, kind="bar", height=6, aspect=2, palette="muted", legend=False)
plt.title("Bar Plot: Survival Rate by Class and Sex")
plt.show()
```
