## 1. Google Play Store Data Analysis: Pie and Bar Charts
### 1A. Read and Filter Data. 
##### This snippet reads the "googleplaystore.csv" file into a DataFrame and filters out rows with "Adults only 18+" and "Unrated" in the Content Rating column.It then resets the index and displays the updated counts of content ratings.

```python
import pandas as pd

# Read the CSV file into a DataFrame
pstore = pd.read_csv("googleplaystore.csv")

# Display the original content rating counts
print("Original Content Rating counts:")
print(pstore['Content Rating'].value_counts())

# Remove rows with "Adults only 18+" and "Unrated" ratings
pstore = pstore[~pstore['Content Rating'].isin(["Adults only 18+", "Unrated"])]

# Reset the DataFrame index after filtering
pstore.reset_index(inplace=True, drop=True)

# Display the filtered content rating counts
print("\nFiltered Content Rating counts:")
print(pstore['Content Rating'].value_counts())
```

### 1B. Pie Chart
#### This snippet plots a pie chart of the Content Rating counts from the Google Play Store data. It sets a figure size for better visualization and displays the pie chart.

```python
import matplotlib.pyplot as plt

# Create a pie chart of the Content Rating counts
plt.figure(figsize=[9, 7])
pstore['Content Rating'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Content Rating Distribution (Pie Chart)")
plt.ylabel("")  # Optional: remove y-label for a cleaner look
plt.show()
```

### 1C. Bar Chart
#### This snippet creates a horizontal bar chart of the Content Rating counts. It uses a similar figure size and displays the distribution in bar format for easy comparison.

```python
import matplotlib.pyplot as plt

# Create a horizontal bar chart of the Content Rating counts
plt.figure(figsize=[9, 7])
pstore['Content Rating'].value_counts().plot.barh(color='skyblue')
plt.title("Content Rating Distribution (Bar Chart)")
plt.xlabel("Count")
plt.ylabel("Content Rating")
plt.show()
```

## 2. Scatter Plots Using the Tips Dataset
### 2A. Basic Scatter Plot
#### This snippet loads the "tips" dataset from Seaborn and plots a basic scatter plot with total bill versus tip. It provides a simple visualization of the relationship between total bill and tip amounts.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a basic scatter plot: total_bill vs. tip
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

### 2B. Scatter Plot with Hue (Time)
#### This snippet creates a scatter plot with points colored by the "time" variable. It helps differentiate data points based on whether the meal was lunch or dinner.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a scatter plot with color (hue) based on meal time
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Scatter Plot: Total Bill vs Tip (Hue: Time)")
plt.show()
```

### 2C. Scatter Plot with Hue and Style (Time)
#### This snippet differentiates data points by both color and marker style using the "time" variable. It improves visual distinction between lunch and dinner data points.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a scatter plot with hue and style based on meal time
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="time")
plt.title("Scatter Plot: Total Bill vs Tip (Hue & Style: Time)")
plt.show()
```

### 2D. Scatter Plot with Hue (Day) and Style (Time)
#### This snippet creates a scatter plot using different hues for "day" and marker styles for "time." It allows simultaneous visualization of the impact of both day and time on tipping behavior.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a scatter plot with hue based on day and style based on time
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")
plt.title("Scatter Plot: Total Bill vs Tip (Hue: Day, Style: Time)")
plt.show()
```

### 3A. Basic Pair Plot
#### This snippet loads the Iris dataset and creates a pair plot for all variables. It provides a matrix of scatter plots and histograms to explore pairwise relationships in the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot for the iris dataset
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(iris)
plt.suptitle("Pair Plot: Iris Dataset", y=1.02)
plt.show()
```

### 3B. Pair Plot for Selected Variables
#### This snippet creates a pair plot focusing only on "sepal_width" and "sepal_length." It helps to closely examine the relationship between these two specific features.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot for selected variables
g = sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
plt.suptitle("Pair Plot: Sepal Width and Length", y=1.02)
plt.show()
```

### 3C. Pair Plot with KDE on the Diagonals
#### This snippet creates a pair plot where the diagonal plots show Kernel Density Estimates (KDE). It provides a smooth estimate of the distribution for each variable along the diagonal.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot with KDE on the diagonal
g = sns.pairplot(iris, diag_kind="kde")
plt.suptitle("Pair Plot with KDE Diagonals", y=1.02)
plt.show()
```

## 4. Heatmap Examples
### 4A. Basic Heatmap
#### This snippet creates a basic heatmap using randomly generated data. It sets a fixed range for the color scale and displays the heatmap.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility and generate uniform data
np.random.seed(0)
sns.set()
uniform_data = np.random.rand(10, 12)

# Create a heatmap with a fixed color scale (0 to 1)
ax = sns.heatmap(uniform_data, vmin=0, vmax=1)
plt.title("Basic Heatmap")
plt.show()
```

### 4B. Heatmap with Correlation Matrix and Custom Colors
#### This snippet computes a correlation matrix from random data and plots it as a heatmap.It customizes the heatmap with a "Blues" colormap, removes the color bar, and adds grid lines.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a random DataFrame and compute its correlation matrix
df = pd.DataFrame(np.random.randn(50).reshape(10, 5))
corr = df.corr()

# Plot the correlation matrix as a heatmap with custom settings
ax1 = sns.heatmap(corr, cbar=False, linewidths=2, vmax=1, vmin=0, square=True, cmap='Blues')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

### 4C. Heatmap of Flight Data
#### This snippet loads the flights dataset from Seaborn and reshapes it into a pivot table. It then creates a heatmap to visualize passenger numbers over months and years.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset and pivot the data
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")

# Plot a heatmap of the flight data
ax = sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap: Flight Data")
plt.show()
```
