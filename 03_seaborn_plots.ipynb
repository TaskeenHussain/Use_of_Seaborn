### 1. Importing Libraries and Plotting a Pairplot (Penguins Dataset)
#### This snippet imports libraries, loads the "penguins" dataset from Seaborn, and creates a pairplot with points colored by species.It provides a quick overview of pairwise relationships among variables in the dataset.

### 2. Applying Default Seaborn Theme, Scaling, and Color Palettes
#### This snippet defines a function to plot the current Seaborn color palette and then iterates through a list of predefined palettes.It sets the Seaborn palette and displays the corresponding color palette for visual inspection.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to display the current Seaborn color palette
def plot_color_palette(palette: str):
    sns.palplot(sns.color_palette())
    plt.xlabel("Color palette: " + palette)
    plt.show()

# List of palettes to cycle through
palettes = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]

# Apply each palette and plot its color palette
for palette in palettes:
    sns.set(palette=palette)
    plot_color_palette(palette)
```

### 4. Relational Plot with Faceting, Hue, Style, and Size
#### This snippet creates a relational plot (scatter plot) using the "tips" dataset. It facets the data by 'time', colors points by 'smoker', styles by 'smoker', and sizes by 'size'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset (if not already loaded)
tips = sns.load_dataset("tips")

# Create a relational plot with multiple aesthetics and faceting
sns.relplot(x="total_bill", y="tip", 
            col="time",       # Facets the grid by time (Lunch/Dinner)
            hue="smoker",     # Colors points by smoker status
            style="smoker",   # Uses different markers for smoker status
            size="size",      # Varies point size based on party size
            data=tips)
plt.show()
```

### 5. Linear Model Plot Examples
#### Example 5A: lmplot of Total Bill vs Tip
##### This snippet creates a linear regression plot (lmplot) of total bill versus tip from the "tips" dataset. It shows the relationship with a fitted regression line.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Linear model plot for total_bill vs tip
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.title("Linear Regression: Total Bill vs Tip")
plt.show()
```

### Example 5B: lmplot of Size vs Tip
#### This snippet creates an lmplot to examine the relationship between party size and tip. It provides a regression line to highlight the trend between these variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Linear model plot for size vs tip
sns.lmplot(x="size", y="tip", data=tips)
plt.title("Linear Regression: Party Size vs Tip")
plt.show()
```

### Example 5C: lmplot with Mean Estimator
#### This snippet applies a custom estimator (mean) to the lmplot for size vs tip. It plots the regression line based on the mean values of x (party size) and y (tip).

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# lmplot with a custom x estimator (mean) for size vs tip
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean)
plt.title("lmplot with Mean Estimator: Size vs Tip")
plt.show()
```

### Example 5D: Logistic Regression Plot
#### This snippet creates a logistic regression plot using lmplot for a binary outcome based on tip ratio. It first creates a new column "big_tip" (True if tip/total_bill > 0.15) and then plots the logistic regression curve.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a binary column: True if tip/total_bill > 0.15
tips["big_tip"] = (tips.tip / tips.total_bill) > 0.15

# Create a logistic regression plot with jitter on the y-axis
sns.lmplot(x="total_bill", y="big_tip",
           y_jitter=0.03, 
           logistic=True, 
           data=tips)
plt.title("Logistic Regression: Total Bill vs Big Tip")
plt.show()
```

### Example 5E: Joint Plot with Regression
#### This snippet creates a joint plot with regression (kind="reg") to explore the relationship between total bill and tip. It combines scatter plots with a regression line and histograms for marginal distributions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a joint plot with regression
sns.jointplot(x="total_bill", y="tip", 
              kind="reg",  # Options: "scatter", "reg", "resid", "kde", "hex"
              data=tips)
plt.suptitle("Joint Plot: Total Bill vs Tip", y=1.02)
plt.show()
```

### Example 5F: Pairplot of Tips Data by Smoker

## 6. API Abstraction Across Visualizations
### 6A: Line Plot with Faceting (Dots Dataset)
#### This snippet loads the "dots" dataset and creates a line plot with multiple facets based on the "align" variable. It uses various aesthetics like hue, size, and style to encode additional information.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dots dataset
dots = sns.load_dataset("dots")

# Create a line plot with faceting, using multiple aesthetics
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False)
)
plt.suptitle("Line Plot with Faceting: Dots Dataset", y=1.02)
plt.show()
```

## 7. Statistical Estimation and Error Bars
### 7A: Line Plot with Error Bars (FMRI Dataset)
#### This snippet loads the "fmri" dataset and creates a line plot with multiple facets based on the "region" variable. It shows how different events affect the signal over time, with error bars representing variability.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot with facets for different brain regions and error bars
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event"
)
plt.suptitle("FMRI Signal Over Time by Region", y=1.02)
plt.show()
```

## 8. Specialized Plots for Categorical Data
### 8A: Categorical Swarm Plot
#### This snippet creates a swarm plot (a type of categorical plot) that shows total bill distribution by day and smoker status. It provides a detailed view of individual data points for each category.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a categorical swarm plot for total bill by day, colored by smoker
sns.catplot(x="day", y="total_bill", 
            hue="smoker", kind="swarm", 
            data=tips)
plt.title("Swarm Plot: Total Bill by Day and Smoker")
plt.show()
```

### 8B: Categorical Violin Plot
#### This snippet creates a violin plot for total bill by day, split by smoker status. It combines a kernel density estimate with a boxplot-like summary for each category.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot for total bill by day, split by smoker status
sns.catplot(x="day", y="total_bill", 
            hue="smoker", kind="violin", split=True, 
            data=tips)
plt.title("Violin Plot: Total Bill by Day (Split by Smoker)")
plt.show()
```

### 8C: Categorical Bar Plot
#### This snippet creates a bar plot for total bill by day, with bars colored by smoker status. It provides aggregated means (with confidence intervals) for each category.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot for total bill by day, colored by smoker status
sns.catplot(x="day", y="total_bill", 
            hue="smoker", kind="bar", 
            data=tips)
plt.title("Bar Plot: Total Bill by Day and Smoker")
plt.show()
```

### 9. Opinionated Defaults and Flexible Customization
#### This snippet demonstrates using Seaborn's defaults to create a relational plot with custom aesthetics from the "penguins" dataset. It visualizes the relationship between bill length and bill depth, with hue mapped to body mass.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
df = sns.load_dataset("penguins")

# Create a relational plot with custom aesthetics
sns.relplot(
    data=df,
    x="bill_length_mm", y="bill_depth_mm", 
    hue="body_mass_g"
)
plt.title("Relational Plot: Bill Dimensions vs Body Mass")
plt.show()
```
