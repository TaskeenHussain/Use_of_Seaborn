### Plot Varieties

### 1. Reading the Google Play Store Data
### This snippet reads the "googleplaystore.csv" file into a Pandas DataFrame and displays the first 10 rows. It is used to inspect the dataset before further analysis.

```python
import pandas as pd

# Read the CSV file into a DataFrame
pstore = pd.read_csv("googleplaystore.csv")

# Display the first 10 rows of the DataFrame
print(pstore.head(10))
```

### 2. Basic Distribution Plot for App Ratings
#### This code creates a distribution plot for the 'Rating' column using Seaborn's distplot. It provides a quick look at the distribution shape of the app ratings.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file (if not already read)
pstore = pd.read_csv("googleplaystore.csv")

# Create a basic distribution plot for the 'Rating' column
sns.distplot(pstore.Rating)
plt.title("Distribution of App Ratings")
plt.show()
```

### 3. Distribution Plot with Custom Number of Bins
#### This snippet plots the distribution of app ratings with 20 bins and without the kernel density estimate (KDE). It focuses on the histogram aspect of the distribution to show frequency counts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file (if not already read)
pstore = pd.read_csv("googleplaystore.csv")

# Plot the distribution with 20 bins and no KDE curve
sns.distplot(pstore.Rating, bins=20, kde=False)
plt.title("Distribution of App Ratings (Histogram Only)")
plt.show()
```

### 4. Distribution Plot with Title and Custom Color
#### This code creates a distribution plot for app ratings with 20 bins and a green color. It also adds a custom title (in red, fontsize 20) to the plot for enhanced visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file (if not already read)
pstore = pd.read_csv("googleplaystore.csv")

# Plot the distribution with 20 bins, set color to green, and add a title
sns.distplot(pstore.Rating, bins=20, color="g")
plt.title("Distribution of App Ratings", fontsize=20, color='red')
plt.show()
```

### 5. Styled Distribution Plot with Dark Background
#### This snippet applies a dark background style to the plot and then creates a distribution plot for the 'Rating' column. Using a dark background can improve the visual appeal of graphs for presentations or reports.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the dark background style for the plot
plt.style.use("dark_background")

# Read the CSV file (if not already read)
pstore = pd.read_csv("googleplaystore.csv")

# Create a distribution plot with 20 bins and green color, on a dark background
sns.distplot(pstore.Rating, bins=20, color="g")
plt.title("Distribution of App Ratings", fontsize=20, color='red')
plt.show()
```
